# Data Visualization
This parts contains the visualization graphs from the data collected.\
For each news account, a bar plot was used to compared the relative level of engagement of tweets for countries with different income levels. 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
"""
Read CSV files that contains data from the data collection process
CSV files have three columns:
1. 'Countries', 
2. '# of Reactions' (mean number of engagement for each country), and
3. 'Income Level'

"""

country_bbc = pd.read_csv("BBC Countries.csv")
country_cnn = pd.read_csv("CNN Countries.csv")
country_reu = pd.read_csv("Reuters Countries.csv")
```

## Reuters Visualization
This contains visualization of data collected from Reuters

 #### Prepare the data for visualization

```python
"""
This extracts different parts of the dataframe and groups countries with the same level of income together. 
It also includes individual rows for the UK, US, and China.

"""

high_reu = country_reu[country_reu['Income Level'] == "High income"]
upper_mid_reu = country_reu[country_reu['Income Level'] == "Upper middle income"]
lower_mid_reu = country_reu[country_reu['Income Level'] == "Lower middle income"]
low_reu = country_reu[country_reu['Income Level'] == "Low income"]

uk_reu = country_reu[country_reu["Countries"]== "United Kingdom"]
us_reu = country_reu[country_reu["Countries"]== "United States"]
china_reu = country_reu[country_reu["Countries"]== "China"]
```

```python
"""
This finds the relative level of engagements for each category and the three countries. 
Relative level means how the number of engagement is relative to the mean for all the countries in the dataframe. 

We divide the mean level of engagements for each income category to the mean level of engagemetns for all countries.
This allows us to get a benchmark and allow us to compare it better, 
as the raw number based on the tweet might not be a comparable data, 
since Reuters have far less followers, thus, reaction level than BBC and CNN. 

Create a dictionary to store the numbers and see the numbers clearer.

"""
reu_react = {}

all_mean = country_reu["# of Reactions"].mean() #find the mean number of engagements for all countries

#round figures to 2 decimal places.
#divide the mean of each category to all mean 
high_mean = round(high_reu["# of Reactions"].mean()/all_mean,2)
upper_mid_mean = round(upper_mid_reu["# of Reactions"].mean()/all_mean,2)
lower_mid_mean = round(lower_mid_reu["# of Reactions"].mean()/all_mean,2)
low_mean = round(low_reu["# of Reactions"].mean()/all_mean,2)

uk_mean = round(uk_reu["# of Reactions"].mean()/all_mean,2)
us_mean = round(us_reu["# of Reactions"].mean()/all_mean,2)
china_mean = round(china_reu["# of Reactions"].mean()/all_mean,2)

all_mean = all_mean/all_mean #benchmark - all_mean is 1.0

#populate the dictionary
reu_react["All Countries"] = all_mean
reu_react["High income"] = high_mean
reu_react["Upper middle income"] = upper_mid_mean
reu_react['Lower middle income'] = lower_mid_mean
reu_react["Low income"] = low_mean
reu_react["UK"] = uk_mean
reu_react["US"] = us_mean
reu_react["China"] = china_mean
```

```python
reu_react
```

#### Bar Chart

```python
plt.figure(figsize=(50, 20))
keys = reu_react.keys()
values = reu_react.values()
plt.bar(keys, values,color=['black', 'blue', 'purple','cyan', 'red', 'yellow','magenta','green'])
plt.rcParams.update({'font.size': 27})
plt.title('Reuters: Relative level of engagements for countries with different income levels',fontweight='bold',fontsize=50)
plt.xlabel('Income Levels',fontweight='bold', fontsize=30)
plt.ylabel('Level of engagement relative to all tweets',fontweight='bold', fontsize = 30)
plt.axhline(y=1,linewidth=1, linestyle='dashed',lw=5,color='red') #benchmark line for comparison
plt.savefig('Reuters Diagram.png', bbox_inches='tight',facecolor='white') # save graph
plt.show()  
```
